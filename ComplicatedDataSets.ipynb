```python
from pandas.api.types import CategoricalDtype
```

```python
df.cut.head()
```

```python
df.cut = df.cut.astype(CategoricalDtype(ordered = True))
```

```python
df.cut
```

```python
cut_categori = ["Fair", "Good","Very Good", "Premium", "Ideal" ]
```

```python
df.cut = df.cut.astype(CategoricalDtype(categories = cut_categori, ordered = True))
```

```python
df.cut
```
